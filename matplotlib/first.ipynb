####  BAR PLOT

```python

x=["python", "c++", "javascript","c#","html"]
y=[1,2,3,4,5]
z=[4,5,1,3,2]
# plt.bar(x,y,width=0.2,color=["red","blue","green","orange","yellow"],align="center",edgecolor="black",linewidth=5,linestyle=":",alpha=0.6,label="keshav")
p=np.arange(len(x))
p1=[j+0.2 for j in p]
# plt.bar(p,y, color="r",label="pop1" ,width=0.2)
# plt.bar(p1,z, color="g",label="pop2", width=0.2)
plt.legend()
plt.xticks(p+0.1,x,rotation=10)
plt.xlabel("language-->")
plt.ylabel("number-->")
plt.title("popularity",fontsize=15)
plt.show()
```

```python
x=["python", "c++", "javascript","c#","html"]
y=[1,2,3,4,5]
# plt.bar(x, y, hatch='//')
# plt.bar(x, y, hatch='--')
# plt.bar(x, y, hatch='xx')
# plt.bar(x, y, hatch='oo')
plt.grid(color="red", linestyle="--", linewidth=0.5, zorder=1)
plt.bar(x, y, color="skyblue", edgecolor="black", linewidth=1, zorder=2)
plt.xlabel("Language -->")
plt.ylabel("Number -->")
plt.title("Bar Chart with Zorder Example", fontsize=15)

plt.show()
```

```python
x=["python", "c++", "javascript","c#","html"]
y=[1,2,3,4,5]
plt.barh(x,y, color="r",label="pop1")
plt.show()
```

# SCATTER PLOT

```python
x=[1,2,3,45,43]
y=[23,5,60,98,76]
c=[10,20,30,40,99]
plt.scatter(x,y,c=c,cmap="viridis")
t=plt.colorbar()
t.set_label("color bar")
plt.show()
```
